# Word clouds for CyBOK

Load mapping reference in JSON format. Then, use this reference to build documents representative of each knowledge area.

```python
import json

with open('data/reference.json') as json_file:
    term2ka = json.load(json_file)

knowledge_areas = [
    'AAA',
    'AC',
    'AB',
    'C',
    'CI',
    'CPS',
    'DSS',
    'F',
    'FMS',
    'HF',
    'HS',
    'LR',
    'MAT',
    'NS',
    'OSV',
    'PLT',
    'POR',
    'RMG',
    'SOIM',
    'SS',
    'SSL',
    'WAM'
]

ka2doc = {
    'AAA': "",
    'AC': "",
    'AB': "",
    'C': "",
    'CI': "",
    'CPS': "",
    'DSS': "",
    'F': "",
    'FMS': "",
    'HF': "",
    'HS': "",
    'LR': "",
    'MAT': "",
    'NS': "",
    'OSV': "",
    'PLT': "",
    'POR': "",
    'RMG': "",
    'SOIM': "",
    'SS': "",
    'SSL': "",
    'WAM': ""
}

for ka in knowledge_areas:
    for term in term2ka.keys():
        if ka in term2ka[term]:
            ka2doc[ka] += term + " "
```

Generate word clouds for knowledge areas.

```python
from wordcloud import WordCloud

ka2cloud = {
    'AAA': "",
    'AC': "",
    'AB': "",
    'C': "",
    'CI': "",
    'CPS': "",
    'DSS': "",
    'F': "",
    'FMS': "",
    'HF': "",
    'HS': "",
    'LR': "",
    'MAT': "",
    'NS': "",
    'OSV': "",
    'PLT': "",
    'POR': "",
    'RMG': "",
    'SOIM': "",
    'SS': "",
    'SSL': "",
    'WAM': ""
}

for ka in knowledge_areas:
    ka2cloud[ka] = WordCloud(max_font_size=40, background_color="white", max_words=50).generate(ka2doc[ka])
```

Demonstrate plotting of cloud for Web Application and Mobile Security knowledge area.

```python
import matplotlib.pyplot as plt

plt.imshow(ka2cloud['WAM'], interpolation='bilinear')
plt.axis("off")
```

Save images.

```python
for ka in knowledge_areas:
    ka2cloud[ka].to_file('./imgs/' + ka + '.png')
```
