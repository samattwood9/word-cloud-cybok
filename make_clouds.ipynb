# Word clouds for CyBOK

Load mapping reference in JSON format. Then, use this reference to build documents representative of each knowledge area.

```python
import json

with open('data/reference.json') as json_file:
    term2ka = json.load(json_file)

knowledge_areas = [
    'AAA',
    'AC',
    'AB',
    'C',
    'CI',
    'CPS',
    'DSS',
    'F',
    'FMS',
    'HF',
    'HS',
    'LR',
    'MAT',
    'NS',
    'OSV',
    'PLT',
    'POR',
    'RMG',
    'SOIM',
    'SS',
    'SSL',
    'WAM'
]

ka2doc = {
    'AAA': "",
    'AC': "",
    'AB': "",
    'C': "",
    'CI': "",
    'CPS': "",
    'DSS': "",
    'F': "",
    'FMS': "",
    'HF': "",
    'HS': "",
    'LR': "",
    'MAT': "",
    'NS': "",
    'OSV': "",
    'PLT': "",
    'POR': "",
    'RMG': "",
    'SOIM': "",
    'SS': "",
    'SSL': "",
    'WAM': ""
}

for ka in knowledge_areas:
    for term in term2ka.keys():
        if ka in term2ka[term]:
            ka2doc[ka] += term + " "
```

Generate word clouds for knowledge areas.

```python
from wordcloud import WordCloud

ka2cloud = {
    'AAA': "",
    'AC': "",
    'AB': "",
    'C': "",
    'CI': "",
    'CPS': "",
    'DSS': "",
    'F': "",
    'FMS': "",
    'HF': "",
    'HS': "",
    'LR': "",
    'MAT': "",
    'NS': "",
    'OSV': "",
    'PLT': "",
    'POR': "",
    'RMG': "",
    'SOIM': "",
    'SS': "",
    'SSL': "",
    'WAM': ""
}

for ka in knowledge_areas:
    ka2cloud[ka] = WordCloud(max_font_size=40, background_color="white", max_words=50).generate(ka2doc[ka])
```

Demonstrate plotting of cloud for Web Application and Mobile Security knowledge area.

```python
import matplotlib.pyplot as plt

plt.imshow(ka2cloud['WAM'], interpolation='bilinear')
plt.axis("off")
```

Save images.

```python
for ka in knowledge_areas:
    ka2cloud[ka].to_file('./imgs/' + ka + '.png')
```

Prepare to do the same for knowledge groups.

```python
ka2group = {
    'AAA': "Systems",
    'AC': "Infrastructure",
    'AB': "Attacks & Defences",
    'C': "Systems",
    'CI': "Introduction",
    'CPS': "Infrastructure",
    'DSS': "Systems",
    'F': "Attacks & Defences",
    'FMS': "Systems",
    'HF': "Human, Org., & Reg.",
    'HS': "Infrastructure",
    'LR': "Human, Org., & Reg.",
    'MAT': "Attacks & Defences",
    'NS': "Infrastructure",
    'OSV': "Systems",
    'PLT': "Infrastructure",
    'POR': "Human, Org., & Reg.",
    'RMG': "Human, Org., & Reg.",
    'SOIM': "Attacks & Defences",
    'SS': "Software & Platform",
    'SSL': "Software & Platform",
    'WAM': "Software & Platform"
}

group2doc = {
    'Introduction': "",
    'Human, Org., & Reg.': "",
    'Attacks & Defences': "",
    'Systems': "",
    'Software & Platform': "",
    'Infrastructure': ""
}

for ka in knowledge_areas:
    for term in term2ka.keys():
        if ka in term2ka[term]:
            group2doc[ka2group[ka]] += term + " "
```

Generate word clouds for knowledge groups.

```python
group2cloud = {
    'Introduction': "",
    'Human, Org., & Reg.': "",
    'Attacks & Defences': "",
    'Systems': "",
    'Software & Platform': "",
    'Infrastructure': ""
}

for group in group2cloud.keys():
    group2cloud[group] = WordCloud(max_font_size=40, background_color="white", max_words=50).generate(group2doc[group])
```

Show an example for a knowledge group.

```python
plt.imshow(group2cloud['Systems'], interpolation='bilinear')
plt.axis("off")
```

Save images.

```python
for group in group2cloud.keys():
    group2cloud[group].to_file('./imgs/' + group + '.png')
```
